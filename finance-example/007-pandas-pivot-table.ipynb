# pivot-table: 四个参数：index, values, columns, aggfunc

## index: 分类汇总的分类条件


```python
df.pivot_table(index=['Gender', 'Education'])
# 展示的值是mean
```

```python
# values可以用list，来空值输出的数值
df.pivot_table(index=['Gender', 'Marital Status'],values=['Income'])
```

```python
# default，数值是mean。因为默认aggfunc=mean()
df.pivot_table(index=['Gender', 'Marital Status'],values=['Income'],aggfunc=max)
```

```python
# columns 控制输出的columns
pd.pivot_table(df, index=['Gender', 'Education'], columns='Marital Status', values=['Height', 'Weight', 'Income'])
```
