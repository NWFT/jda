```python
import numpy as np

def zscore(x, data):
    """
    Calculate the z-score of a given data point x based on a data set data.
    
    Args:
    x (float): The data point for which to calculate the z-score.
    data (array-like): The data set used to calculate the mean and standard deviation.
    
    Returns:
    float: The z-score of the data point x.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z = (x - mean) / std_dev
    return z
```

```python
data = [0.1,2, 4, 6, 8, 10,100]
x = 107
z = zscore(x, data)
print(z)
```
