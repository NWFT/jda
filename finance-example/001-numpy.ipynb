# numpy Index/Slice

```python
arr = np.random.randint(1,100,size=(5,6))
arr
```

```python
# one line
arr[0]
```

```python
# one element
arr[0,-1]
```

```python
# many selected lines
arr[[0,3,4]]
```

```python
# slice the first 2-lines
arr[0:2]
```

```python
# slice the first 2-columns
arr[:,0:2]
```

```python
arr[:,-1]
```

```python
# 2-rows, 2-columns
arr[0:2,0:2]
```

```python
# reverse array
# 行反转
arr[::-1]
```

```python
# reverse array
# 列反转
arr[:,::-1]
```

```python
# reverse array
# ALL 反转
arr[::-1,::-1]
```

```python
# img processing
img_arr = img_arr[::-1,::-1,::-1]
plt.imshow(img_arr)
```

```python
# img slice
plt.imshow(img_arr[50:250,100:400])
```

# 常用统计函数

```python
 # 一组元素中，最大值-最小值
np.ptp(arr[0])
```

```python
np.ptp(arr,axis=1) # 0-rows; 1-columns
```

```python
# 一定方向的最大最小值 amax（）
np.amin(arr,axis=1)  
```

```python
# Median() 中值(奇数-中间值，偶数-2个中间的平均值)
np.median(arr)  
```

```python
np.median(np.arange(1,7,step=1))
```

```python
# Std() Standard Division. std = sqrt( mean( (x-x.mean())**2 ) )
# 表示 一组数据中，数值的离散程度。数值越大，数值分布越离散
np.std(np.arange(1,7,step=1))
arr[1].std()
```

```python
# var()  var = mean( (x-x.mean())**2 )
arr[1].var()
```
