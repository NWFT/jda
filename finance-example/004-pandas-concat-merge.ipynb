# concat, append

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame(data=np.random.randint(0,10,size=(7,3)),columns=["A","B","C"])
df2 = pd.DataFrame(data=np.random.randint(0,10,size=(7,3)),columns=["A","D","C"])
```

```python
pd.concat((df1,df2),axis=1)
```

```python
# Join: Union
pd.concat((df1,df2),axis=0)
```

```python
# Join: inner
pd.concat((df1,df2),axis=0,join='inner')
```

# merge合并：是对表内数据的操作
# concat：会改变表结构

```python
df1 = pd.DataFrame({
    'name':['Lisa','Bob','Jakie'],
    'dept':['hr','accounting','rd']
})

df2 = pd.DataFrame({
    'name':['Jakie','Bob','Lisa'],
    'join':['2020','2021','2000']
})
```

```python
pd.merge(df1,df2,on='name')
```

```python
df3 = pd.DataFrame({
    'dept':['hr','accounting','engineer'],
    'city':['Vancouver','Toranto','Calgray']
})
```

```python
pd.merge(df1,df3)
```

```python
pd.merge(df1,df3, how='outer')  # how-> 'outer', 'inner', 'left', 'right'
```

```python
pd.merge(df1,df3, how='left')
```

```python
pd.merge(df1,df3, how='right')
```

```python
# 如果2个df，column名字不一致
pd.merge(df2, df3, left_on='name', right_on='dept')
```
