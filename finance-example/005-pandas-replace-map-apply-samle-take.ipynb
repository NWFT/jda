# map (是Series方法)  把Series中的每个元素依次带入运算

```python
df = pd.DataFrame({
    'name': ["ZhangSan","LiSi","WangWu"],
    'salary': [1500,2000,4000]
})

dic = {
    "ZhangSan": 'Tom',
    "LiSi": 'Jack'
}
```

```python
df['e_name'] = df['name'].map(dic)
```

```python
df
```

# map使用自定义函数

```python
def neat_salary(x):
    if x > 3000:
        return (x - (x-3000)*0.5)
    else:
        return x
df['n_salary'] = df['salary'].map(neat_salary)
```

```python
df
```

```python
df['salary'].map(neat_salary)
```

```python
df.dtypes
```

# apply 可以用到dataframe，不仅Series

```python
df[['salary','n_salary']].apply(lambda x: x+1000)
```

# 随机抽样

```python
df = pd.DataFrame(np.random.randint(1,100,size=(100,3)),columns=['A','B','C'])
```

```python
df
```

```python
np.random.permutation(3)
```

```python
df.take([2,0,1],axis=1)
df.take(np.random.permutation(3),axis=1)
```

```python
# 把所有 cloumns 和 rows，都打乱
df.take(np.random.permutation(3),axis=1).take(np.random.permutation(100),axis=0)
```

```python
# 把前12行，打乱为随机顺序
df.take(np.random.permutation(12))
```

```python
# 从所有数值，随机抽取
df.sample(10)
```
